```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 微软雅黑
plt.rcParams['axes.unicode_minus'] = False

# 设置随机种子以保证结果可复现
np.random.seed(42)

# 1. 生成随机日期和销售数据
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = [start_date + timedelta(days=x) for x in range(0, (end_date - start_date).days + 1, 7)]  # 每周一个数据点

# 生成随机销售数据 (50-200之间)
sales_data = np.random.randint(50, 200, size=len(date_range))

# 2. 创建DataFrame
sales_df = pd.DataFrame({
    'Date': date_range,
    'Sales': sales_data
})

# 3. 绘制图像
plt.figure(figsize=(12, 6))

# 绘制折线图
plt.plot(sales_df['Date'], sales_df['Sales'],
         marker='o', linestyle='-', color='b', linewidth=2, markersize=6)

# 添加标题和标签
plt.title('2023年每周销售数据', fontsize=16, pad=20)
plt.xlabel('日期', fontsize=12)
plt.ylabel('销售额', fontsize=12)

# 旋转x轴标签以便更好地显示日期
plt.xticks(rotation=45)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.6)

# 调整布局防止标签被截断
plt.tight_layout()

# 显示图像
plt.show()

# 4. 可选：保存图像
plt.savefig('sales_plot.png', dpi=300, bbox_inches='tight')
print("图像已保存为sales_plot.png")
```
